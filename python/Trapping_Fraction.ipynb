# Magnetic Trap Acceptance vs. Depth

The minimum emission angle for electrons to be trapped is given by the following expression (see, e.g. Ashtari Esfahani et. al. (2019))

\begin{align*}
\theta_{bot} \ge \sin^{-1}\left(\sqrt{1-\frac{\Delta B}{B_{max}}}\right)\; ,
\end{align*}

where $\Delta B$ is the trap depth and $B_{max}$ is the maximum field strength. For now we will take $B_{max} = 1$ Tesla to be the background CRES field strength. The cosine of this angle represents the corresponding fraction of solid angle, and in the case of an isotropic decay distribution, it also represents the fraction of decay electrons that are trapped.

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
matplotlib.rc('xtick', labelsize=12) 
matplotlib.rc('ytick', labelsize=12) 
matplotlib.rc('axes',  labelsize=12) 
matplotlib.rc('figure', titlesize=12) 
```

```python
def emission_angle_fraction(delta_b):
    # Note that delta_b is in mT:
    theta_min = np.arcsin(np.sqrt(1.-(delta_b/1000.)))
    # return theta_min*180./np.pi
    return np.cos(theta_min)
```

```python
print("fraction at 4.9mT = ",emission_angle_fraction(0.0049))

# Make an array of trap depths:
b = np.arange(0.0,50,0.01)
f = emission_angle_fraction(b)


fig, ax = plt.subplots()
fig.set_size_inches(12, 8)
ax.plot(b, f)

ax.set_xticks(np.arange(0.0,60,5))
ax.set_yticks(np.arange(0.0,0.25,0.02))
ax.set(xlabel='Trap Depth (mT)', ylabel='Fraction of electrons trapped',
       title='Fraction of electrons trapped vs. trap depth for a 1 Tesla CRES field')
ax.grid()
```
